```python
class node:
    val = -1
    A = ""
    lNode = None
    rNode = None
    def __init(A, lNode, rNode):
        self.A = A
        self.lNode = lNode
        self.rNode = rNode
    
    def __init__(val):
        self.val = val
    
    def getVal():
        return val
    
def var_impurity(K, K_0, K_1):
    return (K_0 * K_1) / K**2

def inf_gain(H, X_col, S, tA):
    sum = 0
    K = len(S)
    for x in np.unique(S[X_col]):
        S_n = S[S[X_col] == x]
        K_v = len(S_n)
        K_0 = len(S_n[S_n[X_col] != x])
        K_1 = len(S_n[S_n[X_col] == x])
        sum = sum + (K_v / K) * var_impurity(K_v, K_0, K_1)
    return H - sum;

#Take in data S and target attribute tA
def GrowTree(S, tA):
    #Calculate entropy / variance impurity
    H = var_impurity(len(S), len(S[S[tA] == 0]), len(S[S[tA] == 1]))
    if ( all(x == 0 for x in list(S[tA])) ):
        return S
    elif ( all(x == 1 for x in list(S[tA])) ):
        return S
    else:
        max_gain = (0, 0)
        for col in S.drop(columns=[tA]):
            new_gain = inf_gain(H, col, S, tA)
            max_gain = (new_gain > max_gain[1]) if (col, new_gain) else max_gain
        x_j = max_gain[0]
        return x_j
        S_0 = S[S[x_j] == 0].reset_index()
        S_1 = S[S[x_j] == 1].reset_index()
        return node(x_j, GrowTree(S_0, tA), GrowTree(S_1, tA))
```
